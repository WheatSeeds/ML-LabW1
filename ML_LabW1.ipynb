### 1. Импортируйте необходимые библиотеки

```python
import pandas as pd
import numpy as np
```

### 2. Импортируйте набор данных из файла salary.csv и поместите его в датафрейм

```python
df = pd.read_csv("salary.csv")
```

### 3. Выведите 10 первых строк

```python
print(df.head(10))
```

### 4. Сколько строк и столбцов в наборе данных?

```python
print("Количество строк:", df.shape[0])
print("Количество столбцов:", df.shape[1])
```

### 5. Проверьте набор данных на наличие отсутствующих значений и дубликатов строк. При наличии обработайте их тем или иным способом.

```python
print("Пропущенные значения: ", df.isnull().sum().sum())
print("Дубликаты: ", df.duplicated().sum())
df = df.drop_duplicates()
df = df.ffill()
```

### 6. Получите сводную информацию по датафрейму

```python
df.info()
```

### 7. Что является индексом этого набора данных? Проиндексируйте набор данных по полю age

```python
print("Изначальные индексы набора данных: ", df.index)
df = df.set_index('age')
print("Индексы после индексирования по полю age: ", df.index)
```

### 8. Сколько видов деятельности упоминается в наборе данных? Перечислите их

```python
occupation_count = df['occupation'].nunique()
print("Количество видов деятельности: ", occupation_count)
print("Виды деятельности:")
print(df['occupation'].unique())
```

### 9. Разделите набор данных на два датафрейма по размеру заработной платы (<=50 тыс. и >50 тыс.)

```python
df_low = df[df['salary'] == ' <=50K']
df_high = df[df['salary'] == ' >50K']
print("Зарплата <=50K:", df_low.shape[0], "записей")
print("Зарплата >50K:", df_high.shape[0], "записей")
```

### 10.Сколько черных чиновников, работающих в правительстве разного уровня, имеют положительный прирост капитала?

```python
gov_black_capital_gain = df[(df['race'] == ' Black') &
                           (df['workclass'].str.contains('gov', case=False)) &
                           (df['capital-gain'] > 0)].shape[0]
print("Черных чиновников с положительным приростом капитала: ", gov_black_capital_gain)
```

### 11.Влияет ли семейное положение работника на размер заработной платы?

```python
print("Влияние семейного положения на зарплату:")
marital_salary = pd.crosstab(df['marital-status'], df['salary'], normalize='index') * 100
print(marital_salary.round(1))
```

### 12.Влияет ли образование работника на размер заработной платы?

```python
print("Влияние образования на зарплату:")
education_salary = pd.crosstab(df['education'], df['salary'], normalize='index') * 100
print(education_salary.round(1))
```

### 13.Связано ли число часов, отработанных в неделю с размером заработной платы?

```python
print("Связь часов в неделю и зарплаты:")
hours_salary = df.groupby('salary')['hours-per-week'].describe()
print(hours_salary)
```

### 14.Имеется ли расовое неравенство при получении работниками академической степени бакалавра и магистра?

```python
print("Расовое распределение среди бакалавров и магистров:")
higher_education = df[df['education'].isin([' Bachelors', ' Masters'])]
race_education = pd.crosstab(higher_education['race'], higher_education['education'])
print(race_education)
```

### 15.В каком виде деятельности преимущественно специализируются выходцы из Польши и каким образованием они обладают?

```python
poland_occupation = df[df['native-country'] == ' Poland']['occupation'].value_counts()
poland_education = df[df['native-country'] == ' Poland']['education'].value_counts()

print("Выходцы из Польши - виды деятельности:")
print(poland_occupation)
print("Выходцы из Польши - образование:")
print(poland_education)
```
